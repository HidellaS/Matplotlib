# Subplots 

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = x*2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

# Colours and Styling

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label='X vs X')
ax.plot(x, x**2, label='X to the power 2')
ax.legend()
```

```python
x = np.linspace(0, 11, 10)
```

```python
plt.plot(x, x, color='#bf42f5')
plt.plot(x, x**2, color='#f54242')
plt.plot(x, x + 2, color='#6cf542')
```
